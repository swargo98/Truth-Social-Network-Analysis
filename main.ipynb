Making a table named " truth_and_url" joining "truth_external_url_edges" and "external_urls" with following columns:
truth_id , url_id, url

```python
truth_and_url = truth_external_url_edges.merge(external_urls, left_on="url_id", right_on="id", suffixes=('', '_y'))

# Select the desired columns
truth_and_url = truth_and_url[['truth_id', 'url_id', 'url']]

print(truth_and_url)
```

Finding author id joining previous table and "truths" table   to form a new table "truthID_authorId_url" with following columns:
truth_id, author_id, url,

```python
# Merge last table with `truth` to add the author_id
truthID_authorId_url = truth_and_url.merge(truths[['id', 'author']], left_on="truth_id", right_on="id", suffixes=('', '_y'))

# Select only the desired columns
truthID_authorId_url = truthID_authorId_url[['truth_id', 'author',  'url']]

# Rename columns for clarity
truthID_authorId_url = truthID_authorId_url.rename(columns={'author': 'author_id', 'url': 'url'})

print(truthID_authorId_url)
```

Join previous table with "users" table to get the username for corresponding user_id.
new table - "truth_username_url"
columns -truth_id,user_id,username,url

```python
# Merge last table with `authors` to add the username
truth_username_url = truthID_authorId_url.merge(users[['id', 'username']], left_on="author_id", right_on="id", suffixes=('', '_y'))

# Select only the desired columns
truth_username_url = truth_username_url[['truth_id', 'author_id', 'username', 'url']]

# Rename 'author_id' to 'user_id' for clarity
truth_username_url = truth_username_url.rename(columns={'author_id': 'user_id'})

print(truth_username_url)
```

Checking if any NaN values. But there is no. Which means joining was perfect!

```python
# Count NaN values in each column of merged_result
nan_counts = truth_username_url.isna().sum()

print(nan_counts)
```

```python
# join the user_id frequency of left, right and neutral users
user_freq = left_user_freq.merge(right_user_freq, on='User', how='outer', suffixes=('_left', '_right'))
user_freq = user_freq.merge(neutral_user_freq, on='User', how='outer')

# fill NaN values with 0
user_freq = user_freq.fillna(0)

user_freq[['TruthCount_left', 'TruthCount_right', 'TruthCount_neutral']] = user_freq[['TruthCount_left', 'TruthCount_right', 'TruthCount_neutral']].astype(int)

user_freq
```

```python
# Initialize the 'Political Leaning' column with a default value (e.g., 'Undetermined')
user_freq['Political Leaning'] = 'Undetermined'

# Iterate over each row to determine the political leaning
for index, row in user_freq.iterrows():
    # Compare truth counts and assign the political leaning
    if row['TruthCount_neutral'] > row['TruthCount_left'] and row['TruthCount_neutral'] > row['TruthCount_right']:
        user_freq.at[index, 'Political Leaning'] = 'Neutral'
    elif row['TruthCount_left'] > row['TruthCount_right'] and row['TruthCount_left'] > row['TruthCount_neutral']:
        user_freq.at[index, 'Political Leaning'] = 'Left'
    elif row['TruthCount_right'] > row['TruthCount_left'] and row['TruthCount_right'] > row['TruthCount_neutral']:
        user_freq.at[index, 'Political Leaning'] = 'Right'
    elif row['TruthCount_right'] == row['TruthCount_left'] and row['TruthCount_right'] == row['TruthCount_neutral']:
        user_freq.at[index, 'Political Leaning'] = 'Neutral'
    elif row['TruthCount_right'] == row['TruthCount_left']:
        user_freq.at[index, 'Political Leaning'] = 'Neutral'
    elif row['TruthCount_right'] == row['TruthCount_neutral']:
        user_freq.at[index, 'Political Leaning'] = 'Right'
    elif row['TruthCount_left'] == row['TruthCount_neutral']:
        user_freq.at[index, 'Political Leaning'] = 'Left'
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Initialize a directed graph
G = nx.DiGraph()

# Add nodes with attributes (political leaning)
for _, row in following.iterrows():
    if not G.has_node(row['follower']):
        G.add_node(row['follower'], political_leaning=row['Follower Political Leaning'])
    if not G.has_node(row['followed']):
        G.add_node(row['followed'], political_leaning=row['Followed Political Leaning'])
    G.add_edge(row['follower'], row['followed'])

# Separate nodes by their political leaning
left_nodes = [node for node, attr in G.nodes(data=True) if attr['political_leaning'] == 'Left']
right_nodes = [node for node, attr in G.nodes(data=True) if attr['political_leaning'] == 'Right']
neutral_nodes = [node for node, attr in G.nodes(data=True) if attr['political_leaning'] == 'Neutral']

# Create positions for nodes in columns
pos = {}
x_spacing = 10  # Spacing between columns
y_spacing = 1   # Spacing between nodes in the same column

# Assign positions for 'Left' nodes
for i, node in enumerate(left_nodes):
    pos[node] = (-x_spacing, i * y_spacing * 90)

# Assign positions for 'Neutral' nodes
for i, node in enumerate(neutral_nodes):
    pos[node] = (0, i * y_spacing * 45)

# Assign positions for 'Right' nodes
for i, node in enumerate(right_nodes):
    pos[node] = (x_spacing, i * y_spacing)

# Assign colors to nodes based on their political leaning
color_map = {
    'Left': 'blue',
    'Right': 'red',
    'Neutral': 'green'
}

# Define edge colors based on source and destination node political leaning
edge_colors = []
edge_color_map = {
    ('Left', 'Left'): 'blue',
    ('Left', 'Neutral'): 'purple',
    ('Left', 'Right'): 'cyan',
    ('Neutral', 'Left'): 'orange',
    ('Neutral', 'Neutral'): 'green',
    ('Neutral', 'Right'): 'brown',
    ('Right', 'Left'): 'pink',
    ('Right', 'Neutral'): 'gray',
    ('Right', 'Right'): 'red'
}

for u, v in G.edges:
    source_leading = G.nodes[u]['political_leaning']
    target_leading = G.nodes[v]['political_leaning']
    edge_colors.append(edge_color_map[(source_leading, target_leading)])

# Assign node colors based on their political leaning
node_colors = [color_map[G.nodes[node]['political_leaning']] for node in G.nodes]

# Draw the graph
plt.figure(figsize=(15, 10))
nx.draw(
    G,
    pos,
    node_color=node_colors,
    edge_color=edge_colors,
    node_size=50,
    arrowsize=10,
    with_labels=False
)

# Add a legend for node colors
for label, color in color_map.items():
    plt.scatter([], [], color=color, label=f"Node: {label}")
for (source, target), color in edge_color_map.items():
    plt.scatter([], [], color=color, label=f"Edge: {source} → {target}")

plt.legend(loc='upper right', title='Node and Edge Colors', fontsize=8)
plt.title("Social Network Graph Grouped by Political Leaning", fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```
