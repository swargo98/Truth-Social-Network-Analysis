# Load the 4 tables

```python
import pandas as pd

external_urls = pd.read_csv('dataset/external_urls.tsv', sep='\t')

external_urls
```

```python
truth_external_url_edges = pd.read_csv('dataset/truth_external_url_edges.tsv', sep='\t')

truth_external_url_edges
```

```python
import csv
truths = pd.read_csv('dataset/truths.tsv', sep='\t', on_bad_lines='skip', quoting=csv.QUOTE_NONE)

truths
```

```python
users = pd.read_csv('dataset/users.tsv', sep='\t')

users
```

# Join tables to get user and url count

```python
# drop all columns from truths except id and author
truths = truths[['id', 'author']]
truths
```

```python
# drop all columns from users except id and username
users = users[['id', 'username']]
users
```

```python
# perform left join on external_urls and truth_external_url_edges to get all external urls
# and their corresponding truth ids
external_urls_truths = pd.merge(external_urls, truth_external_url_edges, left_on='id', right_on='url_id', how='left')
external_urls_truths = external_urls_truths[['url', 'truth_id']]
external_urls_truths
```

```python
# perform left join on external_urls_truths and truths to get all authors of external urls
external_urls_truths_authors = pd.merge(external_urls_truths, truths, left_on='truth_id', right_on='id', how='left')
external_urls_truths_authors = external_urls_truths_authors[['url', 'author']]
external_urls_truths_authors
```

```python
# find how many NaN values are in the author column; there were some bad rows in the truths.tsv file, so we need to skip them
external_urls_truths_authors['author'].isnull().sum()  
```

```python
# remove rows with NaN values in the author column
external_urls_truths_authors = external_urls_truths_authors.dropna(subset=['author'])
external_urls_truths_authors
```

```python
# extract the domain from the url
from urllib.parse import urlparse
external_urls_truths_authors['domain'] = external_urls_truths_authors['url'].apply(lambda x: urlparse(x).netloc)
external_urls_truths_authors
```

```python
# find the number of unique domains
external_urls_truths_authors['domain'].nunique()
```

```python
# find the number of unique authors
external_urls_truths_authors['author'].nunique()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import bipartite

# Initialize the graph
G = nx.Graph()

# Add nodes with the bipartite attribute
domains = list(external_urls_truths_authors['domain'])
authors = list(external_urls_truths_authors['author'])

G.add_nodes_from(domains, bipartite=0)
G.add_nodes_from(authors, bipartite=1)

# Add edges
edges = external_urls_truths_authors[['domain', 'author']].values
G.add_edges_from(edges)

# Get positions for the bipartite graph layout
plt.figure(figsize=(15, 15))

# Bipartite layout
pos = nx.bipartite_layout(G, domains)

# Draw nodes for each bipartite set with distinct colors
nx.draw_networkx_nodes(G, pos, nodelist=domains, node_color="lightblue", node_size=100, label="Domains")
nx.draw_networkx_nodes(G, pos, nodelist=authors, node_color="lightgreen", node_size=100, label="Authors")

# Draw edges
nx.draw_networkx_edges(G, pos, edge_color="gray", alpha=0.5)

# Add labels
nx.draw_networkx_labels(G, pos, font_size=8)

# Display the legend
plt.legend(["Domains", "Authors"], loc="upper left")
plt.title("Bipartite Graph of Domains and Authors")
plt.show()
```

```python
# Get the top 5 most frequent authors
top_authors = external_urls_truths_authors['author'].value_counts().head(5).index

# Filter the original dataframe to include only rows with the top authors
filtered_df = external_urls_truths_authors[external_urls_truths_authors['author'].isin(top_authors)]

# Initialize the bipartite graph
G = nx.Graph()

# Add nodes with bipartite attribute
domains = list(filtered_df['domain'])
authors = list(filtered_df['author'])

G.add_nodes_from(domains, bipartite=0)
G.add_nodes_from(authors, bipartite=1)

# Add edges
edges = filtered_df[['domain', 'author']].values
G.add_edges_from(edges)

# Plot the graph
plt.figure(figsize=(15, 15))

# Use bipartite layout for clear separation
pos = nx.bipartite_layout(G, domains)

# Draw the nodes with distinct colors for domains and authors
nx.draw_networkx_nodes(G, pos, nodelist=domains, node_color="lightblue", node_size=100, label="Domains")
nx.draw_networkx_nodes(G, pos, nodelist=authors, node_color="lightgreen", node_size=200, label="Top 5 Authors")

# Draw the edges
nx.draw_networkx_edges(G, pos, edge_color="gray", alpha=0.5)

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=8)

# Add a legend for clarity
plt.legend(["Domains", "Top 5 Authors"], loc="upper left")

plt.title("Bipartite Graph of Domains and Top 5 Authors")
plt.show()
```
